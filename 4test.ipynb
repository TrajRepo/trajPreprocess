```python
import pandas as pd
import time
df = pd.read_csv('E:\\opensky\\states_2021-08-30-00.csv')
df  = df[df['onground']== False]
data = pd.DataFrame(df,columns=['time','icao24', 'lat', 'lon', 'baroaltitude'])
data.info()
```

```python
test = data
# 统计缺失数据
null_data = test[test.isnull().values==True]
null_data.info()
# 统计出缺失数据的icao24
null_icao24 = null_data['icao24'].unique()
print(len(null_icao24))
```

```python
# 去除NAN值，只保留不存在缺失值的数据
all_icao24 = data['icao24'].unique()
save_icao24 = [id for id in all_icao24 if id not in null_icao24]
fileter_data = test[test['icao24'].isin(save_icao24)]
fileter_data.info()
```

```python
# 去除NAN
# data.dropna(axis=0, inplace=True)

fileter_data.sort_values(by=[ 'icao24', 'time'], inplace=True)
# 判断空值
fileter_data.isna().sum()
```

```python
fileter_data['time_stamp']=pd.to_datetime(fileter_data['time'],unit='s')
```

```python
test = fileter_data[fileter_data['icao24'] == '002c4c'].iloc[:10]
```

```python
def fulfill(test):
    start = test['time_stamp'].min()
    end = test['time_stamp'].max()
    pdates = pd.date_range(start=start, end=end, freq='10S')
    df_date = test.set_index(test['time_stamp'])
    #设置连续日期索引
    df_date_new = df_date.reindex(pdates)
    df_date_new.drop(columns=['time', 'time_stamp'], inplace=True)
    df_date_new.reset_index(inplace=True)
    df_date_new = df_date_new.rename(columns = {'index':'time'})
    df_date_new['time'] = df_date_new['time'].apply(lambda x:time.mktime(x.timetuple()))
    df_date_new.ffill(inplace=True)
    return df_date_new
```

```python
df_date_new = fulfill(test)
df_date_new.to_csv("test.csv",index=False)
```

```python
fileter_data.info()
```

```python
data = fileter_data.groupby('icao24', as_index=False).apply(fulfill)
data.sort_values(by=[ 'time','icao24'], inplace=True)
data.reset_index(inplace=True, drop=True)
```
